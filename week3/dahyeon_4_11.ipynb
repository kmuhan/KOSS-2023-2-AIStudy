```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#변수 설정
X= np.arange(-1.0, 1.0, 0.2)
Y= np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10)) #출력값 여기에 저장 모두다 0으로 초기화되어있음

w_x = 2.5
w_y = 3.0

bias = 0.1

```

```python
for i in range(10):
    for j in range(10):
        u = X[i]*w_x + Y[j]*w_y +bias #가중치곱하고 편향 더하기

        y = 1/(1+np.exp(-u)) # 시그모이드 함수
        Z[j][i] = y #저장

plt.imshow(Z, "gray", vmin = 0.0, vmax=1.0)
plt.colorbar()
plt.show()

#검은색이 0 뉴런이 흥분하지 않은 상태
#흰색은 1 뉴런이 흥분한 상태
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
#변수 설정
X= np.arange(-1.0, 1.0, 0.2)
Y= np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10)) #출력값 여기에 저장 모두다 0으로 초기화되어있음

w_im = np.array([[4.0,4.0],
                 [4.0,4.0]])
w_mo = np.array([[1.0],
                 [-1.0]])

b_im = np.array([3.0,-3.0])
b_mo = np.array([0.1])
```

```python
def middle_layer(x,w,b):
    u = np.dot(x,w) + b
    return 1/(1+np.exp(-u))

def output_layer(x, w, b):
    u = np.dot(x,w) + b
    return u
```

```python
for i in range(10):
    for j in range(10):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)

        Z[j][i] = out[0] #저장

plt.imshow(Z, "gray", vmin = 0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

<h3>하이퍼볼릭 탄젠트로 바꾸기</h3>

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
#변수 설정
X= np.arange(-1.0, 1.0, 0.1)
Y= np.arange(-1.0, 1.0, 0.1)

w_im = np.array([[1.0,2.0],
                 [2.0,3.0]])
w_mo = np.array([[-1.0, 1.0],
                 [1.0, -1.0]])

b_im = np.array([0.3,-0.3])
b_mo = np.array([0.4, 0.1])
```

```python
def middle_layer(x,w,b):
    u = np.dot(x,w) + b
    return np.tanh(x) #하이퍼볼릭 탄젠트

def output_layer(x,w,b):
    u = np.dot(x,w) + b
    return np.exp(u)/np.sum(np.exp(u))
```

```python
#각 분류결과 저장
x_1 = []
y_1 = []
x_2 = []
y_2 = []
```

```python
for i in range(20):
    for j in range(20):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)

        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
        else:
            x_2.append(X[i])
            y_2.append(Y[j])

plt.scatter(x_1, y_1, marker="+")
plt.scatter(x_2, y_2, marker="o")
plt.show()
```
